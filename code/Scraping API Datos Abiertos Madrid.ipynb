## 1.- Actividades Culturales y de Ocio Municipales en los próximos 100 dias

```
import json
import requests
import subprocess   #Por un motivo que desconozco no me funciona bien el request. Devuelve caracteres raros. Uso esto en su lugar

f = open("act_temp", "w")
subprocess.call(['curl',
    '-L',
    '-X',
    'GET',
    'https://datos.madrid.es/egob/catalogo/206974-0-agenda-eventos-culturales-100.json'], stdout=f)

inputFile = open('act_temp')
data = json.load(inputFile)

cantidad = len (data ['@graph'])

ind=0
filep = open('actividades.csv', 'w')

#Me voy a quedar con los siguientes datos:

#Fecha de inicio
#Fecha de fin
#Título de la actividad
#Descripción de la actividad
#Distrito
#Latitud
#Longitud

print("{};{};{};{};{};{};{};{}".format("id","fecha_inicio", "fecha_fin", "titulo", "descripcion",
                             "distrito", "latitud", "longitud" ), 
                             file=filep)

while (ind<cantidad):
  #Voy a ir comprobando si existen las claves porque he visto que a algunos registros les faltan esos datos
  try:
      fecha_inicio = data ['@graph'][ind]['dtstart']
  except KeyError:
      fecha_inicio = 'null'
      
  try:
      fecha_fin = data ['@graph'][ind]['dtend']
  except KeyError:
      fecha_fin = 'null'
      
  try:
      titulo = data ['@graph'][ind]['title'].replace(';', '')
  except KeyError:
      titulo = 'null'  
      
  try:
      descripcion = data ['@graph'][ind]['description'].replace(';', '')
  except KeyError:
      descripcion = 'null'   
      
  try:
      direccion = data ['@graph'][ind]['address']['district']['@id']
      distrito = direccion.split('/')[-1] 
  except KeyError:
      distrito = 'null' 
      
      
      
  try:
      latitud = data ['@graph'][ind]['location']['latitude']
  except KeyError:
      latitud = 'null'     
      
  try:
      longitud = data ['@graph'][ind]['location']['longitude']
  except KeyError:
      longitud = 'null'    
      
  ind=ind+1      
  print("{};{};{};{};{};{};{};{}".format(ind,fecha_inicio, fecha_fin, titulo, descripcion,
                             distrito, latitud, longitud ), 
                             file=filep)
  
  
  
filep.close()



#data ['@graph'][0]['dtstart']
#data ['@graph'][0]['dtend']
#data ['@graph'][0]['title']
#data ['@graph'][0]['description']
#data ['@graph'][14]['address']['district']['@id']

#data ['@graph'][14]['location']['latitude']
#data ['@graph'][14]['location']['longitude']




```

```
from googleapiclient.discovery import build
gcs_service = build('storage', 'v1')
```

```
from googleapiclient.http import MediaFileUpload

media = MediaFileUpload('actividades.csv', 
                        mimetype='text/plain',
                        resumable=True)

request = gcs_service.objects().insert(bucket='dataproc-fe7b85fd-43dc-4de0-a520-a824fcd432de-europe-west1', 
                                       name='actividades.csv',
                                       media_body=media)

response = None
while response is None:
  # _ is a placeholder for a progress object that we ignore.
  # (Our file is small, so we skip reporting progress.)
  _, response = request.next_chunk()

print('Upload complete')
```

## 2.- Monumentos de la ciudad de Madrid

```
import json
import requests
import subprocess   #Por un motivo que desconozco no me funciona bien el request. Devuelve caracteres raros. Uso esto en su lugar

f = open("mon_temp", "w")
subprocess.call(['curl',
    '-L',
    '-X',
    'GET',
    'https://datos.madrid.es/egob/catalogo/300356-0-monumentos-ciudad-madrid.json'], stdout=f)

inputFile = open('mon_temp')
data = json.load(inputFile)

cantidad = len (data ['@graph'])

ind=0
filep = open('monumentos.csv', 'w')

#Me voy a quedar con los siguientes datos:

#Titulo
#Barrio
#Distrito
#Codigo Postal
#Direccion
#Latitud
#Lontitud
#Descripcion

print("{};{};{};{};{};{};{};{};{}".format("id","Titulo", "Barrio", "Distrito", "C.Postal",
                             "Direccion", "Latitud", "Longitud", "Descripcion" ), 
                             file=filep)

while (ind<cantidad):
  #Voy a ir comprobando si existen las claves porque he visto que a algunos registros les faltan esos datos
  try:
      titulo = data ['@graph'][ind]['title']
  except KeyError:
      titulo = 'null'
      
  try:
      barrio = data ['@graph'][ind]['address']['area']['@id'].split('/')[-1]
  except KeyError:
      barrio = 'null'
      
  try:
      distrito = data ['@graph'][ind]['address']['district']['@id'].split('/')[-1]
  except KeyError:
      distrito = 'null'  
      
  try:
      cpostal = data ['@graph'][ind]['address']['postal-code']
  except KeyError:
      cpostal = 'null'   
      
  try:
      direccion = data ['@graph'][ind]['address']['street-address']
      
  except KeyError:
      direccion = 'null' 
      
      
  try:
      latitud = data ['@graph'][ind]['location']['latitude']
  except KeyError:
      latitud = 'null'     
      
  try:
      longitud = data ['@graph'][ind]['location']['longitude']
  except KeyError:
      longitud = 'null'    
      
  try:
      descripcion = data ['@graph'][ind]['organization']['organization-desc'].replace(';', '')
  except KeyError:
      descripcion = 'null'    
  
  ind=ind+1
  print("{};{};{};{};{};{};{};{};{}".format(ind,titulo, barrio, distrito, cpostal,
                             direccion, latitud, longitud, descripcion ), 
                             file=filep)
  
  
  
filep.close()


#TITULO
#data ['@graph'][0]['title']

#DESCRIPCION
#data ['@graph'][0]['organization']['organization-desc']

#BARRIO
#data ['@graph'][0]['address']['area']['@id'].split('/')[-1]

#DISTRITO
#data ['@graph'][0]['address']['district']['@id'].split('/')[-1]

#CODIGO POSTAL
#data ['@graph'][0]['address']['postal-code']

#DIRECCION
#data ['@graph'][0]['address']['street-address']


#LATITUD
#data ['@graph'][1]['location']['latitude']

#LONGITUD
#data ['@graph'][1]['location']['longitude']
```
